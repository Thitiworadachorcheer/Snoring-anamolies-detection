```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

file_path = 'data/Snoring Dataset/0/0_0.wav'
y, sr = librosa.load(file_path, sr=16000)
```

```python
# --- Logic ---
#test git push
# Predict FeNo % (Airway Inflammation) - วิเคราะห์จากความแหลม/สากของเสียง
sc = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr)[0])
feno_percent = min((sc / 5000) * 100, 100) # ยิ่งสูงยิ่งเสี่ยงอักเสบ

# Predict ODI % (Oxygen Desat) - วิเคราะห์จากช่วงเงียบ (Apnea)
rms = librosa.feature.rms(y=y)[0]
silence_ratio = np.sum(rms < 0.01) / len(rms)
odi_percent = min(silence_ratio * 500, 100) # สัดส่วนความเงียบสะท้อนความเสี่ยงออกซิเจนต่ำ

# Predict Vascular Stress % - วิเคราะห์จากจังหวะการหายใจ (RRV)
os = np.std(librosa.onset.onset_strength(y=y, sr=sr))
stress_percent = min(os * 15, 100) # ความแปรปรวนจังหวะสะท้อนความเครียดหลอดเลือด

print(f"Inflammation: {feno_percent:.1f}% | ODI: {odi_percent:.1f}% | Stress: {stress_percent:.1f}%")
```
